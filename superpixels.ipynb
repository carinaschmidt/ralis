```python
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float

from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
# load image and convert to floating point
data_path = '/mnt/qb/baumgartner/cschmidt77_data/acdc/slices/train/'
image_name = 'pat_1_diag_2_frame_01_slice_4.npy'
img_path = data_path+image_name
```

```python
img_np = np.load(img_path)
```

```python
#apply SLIC and extract segments
segments = slic(img_np, n_segments=100, compactness=0.01, sigma=1)
```

```python
# show the output
fig = plt.figure("superpixels--%d segments"%(80))
ax = fig.add_subplot(1,1,1)
ax.imshow(mark_boundaries(img_np, segments, color=(1,1,1), outline_color=(1,1,1)))
plt.show()
```
