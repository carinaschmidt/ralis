```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
exp_name = 'lala'
root = '/mnt/qb/baumgartner/cschmidt77_data/BraTS2018/TrainingData/'
output_folder = '/mnt/qb/baumgartner/cschmidt77_data/evaluation_BraTS18/' + exp_name

pat3D = 'volumes/test'
pat3D_labels = 'gts/test'
pat3D_dir = os.path.join(root, pat3D)
pat3D_gt_dir = os.path.join(root, pat3D_labels)
```

```python
pat3D_dir
```

```python
def rc_params():
    plt.rc('text', usetex=True)
    plt.rc('font', **{'family': 'serif', 'sans-serif': 'lmodern', 'size': 15})
    plt.rc('axes', **{'titlesize': 15, 'labelsize': 11})
    plt.rc('xtick', **{'labelsize': 11})
    plt.rc('ytick', **{'labelsize': 11})
    plt.rc('legend', **{'fontsize': 14})
    plt.rc('figure', **{'figsize': (10,5)}) #width, heig
```

```python
title = ['T1', 'T1Gd', 'T2','FLAIR']
```

### Visualisation of BraTS slices

```python
for pat in os.listdir(pat3D_dir):
    pat_dir = os.path.join(pat3D_dir, pat)
    gt_dir = os.path.join(pat3D_gt_dir, pat)
    slice_nr = 77
    if 'pat_143' in pat:
        pat = np.load(pat_dir)
        gt = np.load(gt_dir)
        print(pat.shape)
        print(gt.shape)
        #plt.imshow(pat[:,:,78,1], cmap='gray')
        #plt.imshow(gt[:,:,78], cmap='gray')
        fig, axarr = plt.subplots(1,5)
        rc_params()
        for i in range(0,5):
            if i == 4:
                axarr[i].imshow(gt[:,:,slice_nr ],alpha=1, cmap='afmhot')
                axarr[i].imshow(pat[:,:,slice_nr ,1], alpha=0.6, cmap='gray')
                axarr[i].set_title("(e) T1Gd with \n segmentation")
            else:
                axarr[i].imshow(pat[:,:,slice_nr ,i], cmap='gray')
            if i == 0:
                axarr[i].set_title("(a) "+ title[i])
            elif i == 1:
                axarr[i].set_title("(b) "+ title[i])
            elif i == 2:
                axarr[i].set_title("(c) "+ title[i])
            elif i == 3:
                axarr[i].set_title("(d) "+ title[i])
            axarr[i].axis('off')
        #plt.savefig("pat_143_ModAndSeg.pdf")
        plt.savefig("pat_143_ModAndSeg.pdf", bbox_inches = 'tight')
        break
```
