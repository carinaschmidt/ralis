```python
import numpy as np
import pandas as pd
import os
```

```python
# read log file from supervised and create pandas dataframe 
```

```python
# path to logs
#path = '/mnt/qb/baumgartner/cschmidt77_data/exp2_acdc_supervised/'
path = '/mnt/qb/baumgartner/cschmidt77_data/_FINAL_ACDC_exp2_3Devaluation_supervised/'
```

```python
exp_2_acdc_supervised_list = os.listdir(path)
```

```python
data_frame_lists = []
for exp in exp_2_acdc_supervised_list: 
    exp_path = path + exp + '/Mean_3D_DSC.txt'
    if os.path.isfile(exp_path): 
        names = ["Mean DSC", "Background", "RV", "MYO", "LV", "Experiment", "Augmentation", "lr", "seed", "pre-trained", "train set"]
        data = pd.read_csv(exp_path, nrows=1, skiprows=1, names=names, sep ='\t')
        data["Experiment"] = exp
        data = data[["Experiment", "train set", "pre-trained","Augmentation", "lr", "seed", "Mean DSC", "Background", "RV", "MYO", "LV"]]
        if 'RIRD' in exp:
            data['Augmentation'] = 'RI/RD' 
        elif 'CrFlElCo' in exp:
            data['Augmentation'] = 'RI/RD' 
        else:
            data['Augmentation'] = 'standard' 
        if 'lr0.01' in exp:
            data['lr'] = '0.01'
        elif 'lr0.05' in exp:
            data['lr'] = '0.05'
        if 'ImageNet' in exp:
            data["pre-trained"] = 'ImageNet'
        else:
            data["pre-trained"] = '-'
        if 'allPatients' in exp:
            data['train set'] = '50pat'
        else:
            data['train set'] = '10pat'
        data['seed'] = exp.split('_')[-1]
        data_frame_lists.append(data)

frame = pd.concat(data_frame_lists, axis=0, ignore_index=True)
```
