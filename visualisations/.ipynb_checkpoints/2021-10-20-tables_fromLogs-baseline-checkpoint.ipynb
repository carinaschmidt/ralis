```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# read log file from test RALIS and create pandas dataframe 
```

```python
# path to logs
path = '/mnt/qb/baumgartner/cschmidt77_data/exp2_acdc_baselines/'
```

```python
exp_2_acdc_baselines = os.listdir(path)
```

```python
exp_2_acdc_baselines = [exp for exp in exp_2_acdc_baselines if 'baseline' in exp]
```

```python
exp_2_acdc_baselines
```

```python
data_frame_lists = []
for exp in exp_2_acdc_baselines: 
    exp_path = path + exp + '/test_results.txt'
    if os.path.isfile(exp_path): 
        names = ["Val loss", "Val Acc", "Mean DSC", "DSC Background", "DSC RV", "DSC MYO", "DSC LV", "Experiment", "AL-algo", "seed", "budget"]
        data = pd.read_csv(exp_path, nrows=1, skiprows=1, names=names, sep =',')
        data["Experiment"] = exp
        data = data[["Experiment", "AL-algo", "budget", "seed", "Mean DSC", "DSC Background", "DSC RV", "DSC MYO", "DSC LV", "Val loss", "Val Acc"]]
        if 'bald' in exp:
            data['AL-algo'] = 'bald'
        elif 'entropy' in exp:
            data['AL-algo'] = 'entropy'
        elif 'random' in exp:
            data['AL-algo'] = 'random'
        data['budget'] = exp.split('_')[-1]    
        data['seed'] = exp.split('seed')[1][:2]
        data_frame_lists.append(data)

frame = pd.concat(data_frame_lists, axis=0, ignore_index=True)
```

```python
for budget in frame_new['budget'].unique():
    # create data frames for different budgets names e.g. imgStd01_budget_128
    df_img_std01_budget = frame_new.loc[(frame_new['pre-trained'] == 'ImageNet') & (frame_new['Augmentation'] == 'standard') & (frame_new['lr'] == '0.01') & (frame_new['budget'] == budget)]
    globals()['imgStd01_meanstd_budget_%s' % budget] = df_img_std01_budget.groupby(['pre-trained','Augmentation','lr', 'budget']).agg({'Mean DSC': ['mean', 'std'], 'DSC Background': ['mean', 'std'],
                                                          'DSC RV':['mean', 'std'], 'DSC MYO':['mean', 'std'], 'DSC LV':['mean', 'std']})
```
