```python
import numpy as np
import pandas as pd
import os
```

```python
# read log file from supervised and create pandas dataframe 
```

```python
# path to logs
path = '/mnt/qb/baumgartner/cschmidt77_data/exp2_acdc_supervised/'
```

```python
exp_2_acdc_supervised_list = os.listdir(path)
```

```python
data_frame_lists = []
for exp in exp_2_acdc_supervised_list: 
    exp_path = path + exp + '/test_results.txt'
    if os.path.isfile(exp_path): 
        names = ["Val loss", "Val Acc", "Mean DSC", "DSC Background", "DSC RV", "DSC MYO", "DSC LV", "Experiment", "Augmentation", "lr", "seed", "pre-trained"]
        data = pd.read_csv(exp_path, nrows=1, skiprows=1, names=names, sep =',')
        data["Experiment"] = exp
        data = data[["Experiment", "pre-trained","Augmentation", "lr", "seed", "Mean DSC", "DSC Background", "DSC RV", "DSC MYO", "DSC LV", "Val loss", "Val Acc"]]
        if 'GIGD' in exp:
            data['Augmentation'] = 'GI/GD' 
        elif 'CrFlElCo' in exp:
            data['Augmentation'] = 'standard/RI/GD' 
        else:
            data['Augmentation'] = 'standard' 
        if 'lr0.01' in exp:
            data['lr'] = '0.01'
        elif 'lr0.05' in exp:
            data['lr'] = '0.05'
        if 'ImageNet' in exp:
            data["pre-trained"] = 'ImageNet'
        else:
            data["pre-trained"] = '-'
        data['seed'] = exp.split('_')[-1]
        data_frame_lists.append(data)

frame = pd.concat(data_frame_lists, axis=0, ignore_index=True)
```
