```python
exp_2_acdc_baselines = os.listdir(path)
```

```python
exp_2_acdc_baselines = [exp for exp in exp_2_acdc_baselines]
```

```python
exp_2_acdc_baselines
```

```python
data_frame_lists = []
for exp in exp_2_acdc_baselines: 
    exp_path = path + exp + '/Mean_3D_DSC.txt'
    if os.path.isfile(exp_path): 
        names = ["All classes", "background", "RV", "MYO", "LV", "Experiment", "seed", "budget"]
        data = pd.read_csv(exp_path, nrows=1, skiprows=1, names=names, sep ='\t')
        data["Experiment"] = exp
        data = data[["Experiment", "seed", "budget", "All classes", "background", "RV", "MYO", "LV"]]
        if 'ralis' in exp:
            data['al-algo'] = 'ralis'
        else:
            data['al-algo'] = exp.split('_')[3]
        data['budget'] = exp.split('_')[-3]    
        data['seed'] = exp.split('_')[-1]
        data_frame_lists.append(data)

frame = pd.concat(data_frame_lists, axis=0, ignore_index=True)
```
