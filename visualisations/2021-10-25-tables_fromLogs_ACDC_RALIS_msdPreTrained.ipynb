```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# path to logs
path01 = '/mnt/qb/baumgartner/cschmidt77_data/exp2_3Devaluation_RALIS_msdHeart_lr0.01/'
```

```python
path05 = '/mnt/qb/baumgartner/cschmidt77_data/exp2_3Devaluation_RALIS_msdHeart_lr0.05/'
```

```python
exp_2_acdc_ralis_test_01 = os.listdir(path01)
```

```python
exp_2_acdc_ralis_test_05 = os.listdir(path05)
```

```python
exp_2_acdc_ralis_test = [*exp_2_acdc_ralis_test_01, *exp_2_acdc_ralis_test_05]
```

```python
data_frame_lists = []
for exp in exp_2_acdc_ralis_test: 
    exp_path_01 = path01 + exp + '/Mean_3D_DSC.txt'
    exp_path_05 = path05 + exp + '/Mean_3D_DSC.txt'
    if os.path.isfile(exp_path_01):
        exp_path = exp_path_01
    elif os.path.isfile(exp_path_05):
        exp_path = exp_path_05
    else:
        print("path does not exist: ", exp)
        
    if os.path.isfile(exp_path): 
        names = ["All classes", "background", "RV", "MYO", "LV", "Experiment", "Augmentation", "lr", "seed", "budget"]
        data = pd.read_csv(exp_path, nrows=1, skiprows=1, names=names, sep ='\t')
        data["Experiment"] = exp
        data = data[["Experiment", "Augmentation", "lr", "seed", "budget", "All classes", "background", "RV", "MYO", "LV"]]
        if 'RIRD' in exp:
            data['Augmentation'] = 'RI/RD' 
        elif 'std' in exp:
            data['Augmentation'] = 'standard' 
        else:
            print("check for augmentation")
        if 'lr_0.01' in exp:
            data['lr'] = '0.01'
        elif 'lr_0.05' in exp:
            data['lr'] = '0.05'
        else:
            print("check for learning rate")
        if 'ImageNet' in exp:
            data["pre-trained"] = 'ImageNet'
        elif 'msdHeart' in exp:
            data["pre-trained"] = 'MSDHeart'
        else:
            data["pre-trained"] = '-'
        data['budget'] = exp.split('_')[-3]    
        data['seed'] = exp.split('_')[-1]
        data_frame_lists.append(data)

frame = pd.concat(data_frame_lists, axis=0, ignore_index=True)
```
