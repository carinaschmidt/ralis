```python
data_frame_lists = []
for exp in exp_6_brats: 
    exp_path = path + exp + '/Mean_3D_DSC.txt'
    if os.path.isfile(exp_path): 
        if not 'supervised' in exp:
            names = ["All classes", "background", "NCR/NET", "ED", "ET", "Experiment", "seed", "budget"]
            data = pd.read_csv(exp_path, nrows=1, skiprows=1, names=names, sep ='\t')
            data["Experiment"] = exp
            data = data[["Experiment", "seed", "budget", "All classes", "background", "NCR/NET", "ED", "ET"]]
    #         if 'supervised' in exp:
    #             data['al-algo'] = 'supervised'
    #             data['budget'] = '-'
    #         else:
            data['budget'] = exp.split('_')[-3]    
            if 'ralis' in exp:
                data['al-algo'] = 'ralis'
            elif 'bald' in exp:
                data['al-algo'] = 'bald'
            elif 'random' in exp:
                data['al-algo'] = 'random'
            elif 'entropy' in exp:
                data['al-algo'] = 'entropy'
            else: #supervised all patients
                print("supervised or something went wrong")       
            data['seed'] = exp.split('_')[-1]
            data_frame_lists.append(data)

frame = pd.concat(data_frame_lists, axis=0, ignore_index=True)
```
