```python
def rc_params():
    plt.rc('text', usetex=True)
    plt.rc('font', **{'family': 'serif', 'sans-serif': 'lmodern', 'size': 15})
    plt.rc('axes', **{'titlesize': 15, 'labelsize': 11})
    plt.rc('xtick', **{'labelsize': 11})
    plt.rc('ytick', **{'labelsize': 11})
    plt.rc('legend', **{'fontsize': 14})
    plt.rc('figure', **{'figsize': (8,6)}) #width, heig
```

```python
rc_params()
fig, ax = plt.subplots(2 , 3, figsize=(8,6))
plt.rcParams['font.size'] = '12'
# ACDC
ax[0,0].imshow(slice_npy, cmap='gray') #slice
ax[0,0].set_title("ACDC slice")
ax[0,0].axis('off')
ax[0,1].imshow(pred[0,:,:],cmap='afmhot') #prediction
ax[0,1].set_title("ACDC prediction")
ax[0,1].axis('off')
ax[0,2].imshow(ent[0,:,:], cmap='afmhot') #entropy
ax[0,2].set_title("ACDC entropy map")
ax[0,2].axis('off')
# BraTS
ax[1,0].imshow(brats_slice[:,:,110,1], cmap='gray') #slice
ax[1,0].set_title("BraTS2018 slice T1Gd")
ax[1,0].axis('off')
ax[1,1].imshow(mask_brats[:,:,110,0],cmap='afmhot') #prediction
ax[1,1].set_title("BraTS2018 prediction")
ax[1,1].axis('off')
ax[1,2].imshow(brats_ent[110,0,:,:], cmap='afmhot') #entropy
ax[1,2].set_title("BraTS2018 entropy map")
ax[1,2].axis('off')
plt.tight_layout()
plt.savefig('/home/carina/entropy_ds/acdc_brats_entropy.pdf')
plt.show()
#plt.savefig(os.path.join('/home/carina/entropy_ds/acdc_brats.pdf'))
```
